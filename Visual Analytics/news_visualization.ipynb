```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the first chart: Generational Divide on Gender Pronouns in Work Emails
generations = ['Millennials (25-44)', 'Gen Z (18-24)', 'Age 45+']
support_percentages = [60, 40, 30]
oppose_percentages = [33, 50, 50]

# Bar Chart: Generational Divide on Gender Pronouns in Work Emails
x = np.arange(len(generations))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, support_percentages, width, label='Support', color='skyblue')
bars2 = ax.bar(x + width/2, oppose_percentages, width, label='Oppose', color='salmon')

ax.set_title('Generational Divide on Gender Pronouns in Work Emails', fontsize=14)
ax.set_ylabel('Percentage', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(generations, fontsize=10)
ax.legend(fontsize=10)
ax.bar_label(bars1, fmt='%.0f%%', padding=3, fontsize=10)
ax.bar_label(bars2, fmt='%.0f%%', padding=3, fontsize=10)
plt.tight_layout()

# Save the first chart as an image
plt.savefig("gender_pronouns_email_opinions.png")
plt.show()



# Data for the second chart: Familiarity with Gender-Neutral Pronouns
age_groups = ['18-29', '30-49', '50+']
know_someone = [33, 19, 10]  # Know someone using gender-neutral pronouns
heard_of_someone = [73, 65, 50]  # Heard of someone using gender-neutral pronouns

# Stacked Bar Chart: Familiarity with Gender-Neutral Pronouns
fig, ax = plt.subplots(figsize=(10, 6))
bars_know = ax.bar(age_groups, know_someone, label='Know Someone', color='mediumpurple')
bars_heard = ax.bar(age_groups, heard_of_someone, bottom=know_someone, label='Heard of Someone', color='plum')

ax.set_title('Familiarity with Gender-Neutral Pronouns by Age Group', fontsize=14)
ax.set_ylabel('Percentage', fontsize=12)
ax.legend(fontsize=10)
ax.bar_label(bars_know, fmt='%.0f%%', padding=3, fontsize=10)
ax.bar_label(bars_heard, fmt='%.0f%%', padding=3, fontsize=10)
plt.tight_layout()

# Save the second chart as an image
plt.savefig("gender_neutral_pronouns_familiarity.png")
plt.show()
```
